### 8. Visualize how popularity of genres has changed over the years 

Note that you need to use the number of releases in a year as a parameter of popularity of a genre

Hint 

1: you need to reach to a data frame where the release year is the index and the genre is the column names (one cell shows the number of release in a year in one genre) or vice versa.
Once that is achieved, you can either use univariate plots or can use the heatmap to visualise all the changes over the years 
in one go. 

Hint 2: Use groupby on the relevant column and use sum() on the same to find out the nuumber of releases in a year/genre.  

```python

df_itemcopy=df_item.copy()
df_itemcopy.drop(['movie id', 'movie title', 'release date'],axis=1, inplace=True)
df_itemcopy= df_itemcopy.set_index('release year')
groupped_year = df_itemcopy.groupby(by=['release year']).sum()
groupped_year.sort_values(by='release year', inplace= True)
```

```python
#heatmap using release year and populariry of genre
fig = plt.gcf()  
figsize = fig.get_size_inches()
fig.set_size_inches(figsize * 4.5)  # scale current size by 5.5

sns.heatmap(groupped_year.T,cmap="YlGnBu", annot=True,linewidths=1,linecolor= 'grey')
```

```python
#heatmap using correlation
groupped_yr_corr=groupped_year.corr()
fig=plt.figure(figsize=(25,12))
sns.heatmap(groupped_yr_corr, annot=True,linewidths=0.5, linecolor='white')
```

```python
# plot for year/genre using bar type
# as a example it is plotted for some years. for complete year it will be very big so plotted for few years
groupped_year.iloc[:8,:].plot(figsize=(15, 10), kind='bar')
```

### 10. See gender distribution across different genres check for the validity of the below statements 

* Men watch more drama than women
* Women watch more Sci-Fi than men
* Men watch more Romance than women


1. There is no need to conduct statistical tests around this. Just compare the percentages and comment on the validity of the above statements.

2. you might want ot use the .sum(), .div() function here.
3. Use number of ratings to validate the numbers. For example, if out of 4000 ratings received by women, 3000 are for drama, we will assume that 75% of the women watch drama.

```python
df_user_data= pd.merge(df_user, df_data, how='inner',on='user id')

df_user_data_item=pd.merge(df_user_data,df_item, how='inner',on='movie id')

df_user_data_item.drop(['user id','age','occupation','zip code',
                             'movie id','release date','timestamp','movie title','rating'],axis=1,inplace=True)

df_groupped_gender=df_user_data_item.groupby(by=['gender']).sum()

# dropped some genre which was not required for analysis

df_groupped_gender.drop(['unknown','Action','Adventure','Animation',
        'Childrens', 'Comedy', 'Crime', 'Documentary', 'Fantasy',
        'Film-Noir', 'Horror', 'Musical', 'Mystery',
        'Thriller', 'War', 'Western'],axis=1, inplace=True)

df_groupped_gender
```
